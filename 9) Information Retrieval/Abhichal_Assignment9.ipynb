i. Tokenize the data into words.

ii. Remove all tokens which are stopwords or non alphanumeric characters.

```python
# importing nltk.corpus to import the stopwords
from nltk.corpus import stopwords

# storing the stopwards in a list
stopword=set(stopwords.words('english'))

# variable which hold the clean file
clean_text=[]

# iterating in each word in text
for i in text:
    # checking if i not in stop words
    if i not in stopword:
        # checking if i is alpha numeric
        if i.isalnum():
            # append the word into cleaned text
            clean_text.append(i)
print(clean_text)
```

iii. Convert all words to lowercase.

```python
# iterating over each value in the clean_text
for i in range(len(clean_text)):
    # checking if word is alphabet
    if clean_text[i].isalpha():
        # convert it to lowercase
        clean_text[i]=clean_text[i].lower()
#print(len(clean_text))
print(clean_text)
```

iv. Apply and compare Porter Stemmer and Lancaster Stemmer on top 50 most frequent words. Choose the one that better suits your data and apply on the entire dataset.

```python
# firstly finding the frequency distribution of the words
frequency_distribution=FreqDist(clean_text)
print(frequency_distribution)
```

```python
#finding the top 50 most frequent words
top_50_frequent_words=frequency_distribution.most_common(50)

# list to store only the words without count
freq_words=[]

# iteratinf over each value
for i in top_50_frequent_words:
    #storing the value at index 0 which is the required word
    freq_words.append(i[0])

print(top_50_frequent_words)
```

```python
# importing nltm.stem for applying porterstemmer 
from nltk.stem import PorterStemmer

# initialized an object
portstem=PorterStemmer()

# words after portstemming
portstemmed_words=[]

# iterating over each all the top 50 words
for i in freq_words:
    # storing the portstemmed result after conversion
    portstemmed_words.append(portstem.stem(i))
    

print(portstemmed_words)
```

```python
# importing nltm.stem for applying lancasterstemmer
from nltk.stem import LancasterStemmer

# initialized an object
lancstem=LancasterStemmer()

# words after lancstemming
lancstemmed_words=[]

# iterating over each all the top 50 words
for i in freq_words:
    # storing the lancstemmed result after conversion
    lancstemmed_words.append(lancstem.stem(i))

    
print(lancstemmed_words)
```

# Q2) From Gutenberg project, read the 'blake-poems.txt' as raw data. Perform following operations on the same:

i. Tokenize the data into sentences.

```python
# importing nltk. corpus for importing gutenberg for the text file we need
from nltk.corpus import gutenberg

# storing raw data from the txt file in a variable
data=gutenberg.raw('blake-poems.txt')

# sentence tokenizing the text
set_token=sent_tokenize(data)

set_token
```

ii. For each tokenized sentence remove white-space characters.

```python
# in each line removing the whitespaces
for i in range(len(set_token)):
    # removing the whitespace characters
    set_token[i]=set_token[i].replace('\n',' ')
    set_token[i]=set_token[i].replace('\t',' ')

set_token
```

iii. For each tokenized sentence remove punctuation marks.

```python
# importing nltk for using regextokenizer for removing punctuation marks
import nltk

# for rach sentence
for i in range(len(set_token)):
    
    # creating a tokenizer
    tokenizer = nltk.RegexpTokenizer(r"\w+")
    # passing the sentence in the tokenizer and if punctuation amrks are there they will be removed
    result= tokenizer.tokenize(set_token[i])
    
    # as we are deling with the sentences again connect the words and store them as it is after removing the punctuation
    string=''
    for j in result:
        string+=j+' '
        
    #storing the new sentence at the place of old sentence
    set_token[i]=string
    
    
set_token
```

iv. Convert all words to lower-case.

```python
# iterating on each sentence
for i in range(len(set_token)):
    # used to store the lowercase values
    temp=''
    # for each character in sentence lowercasing it and concatenate with temp
    for j in range(len(set_token[i])):
        temp+=set_token[i][j].lower()
        
    #finally storing the result of temp to the current sentence
    set_token[i]=temp
    
    
set_token
```

v. Remove 'stopwords' from the sentences.

```python
# importing nltk.corpus for using the stopwords
from nltk.corpus import stopwords

# creating a set where ic an store the stopwords
stopword=set(stopwords.words('english'))

# thid will be used to store the clean dataset
clean_data=[]

# for every sentence
for i in set_token:
    # splitting the sentence at the space and storing it in a list
    temp=i.split()
    
    # this will store the sentence free of stopwordss
    ans=''
    # for each word in the temp
    for j in range(len(temp)):
        # if not in stop word adding it to the string
        if temp[j] not in stopword:
            ans+=temp[j]+' '
            
    # appending the cleaned sentence to the clean_data
    clean_data.append(ans)
    
clean_data
```

vi. Remove all non-alpha numeric characters from the sentences.

```python
# iterating over each sentence
for i in clean_data:
    
    # for each character in the sentence
    for j in range(len(i)):
        # if the character is not alphanumeric replace it with empty
        if not i[j].isalnum():
            i[j].replace(i[j],'')
            
clean_data
```
