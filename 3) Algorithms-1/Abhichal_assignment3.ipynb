# 1) IMPLEMENTATION OF SORTING ALGORITHMS

Implement insertion sort, selection sort, bubble sort. Test with arrays of sizes 10, 100, 1000, 10000. Use same array for all sorting algorithms and print time taken by each of them.

```python
# importing all important modules which is useful in program

# importing random module to fenerate random number for list
from random import randint
# inporting timeit module to calculate time taken byb each function
import timeit

# Implementation of bubble sort
def bubblesort(lst):
    # storing length of array
    l=len(lst)
    
    # creating a flag which if not change after one iteration means list sorted and we can break
    flag=0
    for i in range(l-1):
        for j in range(l-i-1):
            #if element greater swap
            if lst[j]>lst[j+1]:
                # swapping two elements
                lst[j],lst[j+1]=lst[j+1],lst[j]
                flag=1
        if flag==0:
            break
    return lst

# Implementation of selection sort
def selectionsort(lst):
    # storing length of array
    l=len(lst)
    for i in range(l):
        # initializing the current value as min
        min=i
        for j in range(i+1,l):
            # compairing if otther value is minimum than the variable min
            if lst[j]<lst[min]:
                # if yes set it as the min
                min=j
        # swapping with minimum
        lst[i],lst[min]=lst[min],lst[i]
    return lst

# Implementation of Insertion Sort
def insertionSort(lst):
    # storing length of array
    l=len(lst)
    for i in range(1,l):
        # storing the current element in the key variable
        key=lst[i]
        index=i-1
        #condition checking of array not go for negative indexing and key less than the current pointing element
        while index>=0 and key<lst[index]:
            lst[index+1]=lst[index]
            index-=1
        lst[index+1]=key
    return lst

# creating four empty lists for 10,100,1000,10000 values repectively
lst1=list()
lst2=list()
lst3=list()
lst4=list()

#generating 10 random integers
for i in range(10):
    lst1.append(randint(1,10))

#generating 100 random integers
for i in range(100):
    lst2.append(randint(1,100))

#generating 1000 random integers
for i in range(1000):
    lst3.append(randint(1,1000))

#generating 10000 random integers
for i in range(10000):
    lst4.append(randint(1,10000))

print('Time Taken by Bubble Sort to sort 10 elements: '+str(timeit.timeit(str(bubblesort(lst1)),'from __main__ import bubblesort,lst1')))
print('Time Taken by Insertion Sort to sort 10 elements: '+str(timeit.timeit(str(insertionSort(lst1)),'from __main__ import insertionSort,lst1')))
print('Time Taken by Selection Sort to sort 10 elements: '+str(timeit.timeit(str(selectionsort(lst1)),'from __main__ import selectionsort,lst1')))
print()
print('Time Taken by Bubble Sort to sort 100 elements: '+str(timeit.timeit(str(bubblesort(lst2)),'from __main__ import bubblesort,lst2')))
print('Time Taken by Insertion Sort to sort 100 elements: '+str(timeit.timeit(str(insertionSort(lst2)),'from __main__ import insertionSort,lst2')))
print('Time Taken by Selection Sort to sort 100 elements: '+str(timeit.timeit(str(selectionsort(lst2)),'from __main__ import selectionsort,lst2')))
print()
print('Time Taken by Bubble Sort to sort 1000 elements: '+str(timeit.timeit(str(bubblesort(lst3)),'from __main__ import bubblesort,lst3')))
print('Time Taken by Insertion Sort to sort 1000 elements: '+str(timeit.timeit(str(insertionSort(lst3)),'from __main__ import insertionSort,lst3')))
print('Time Taken by Selection Sort to sort 1000 elements: '+str(timeit.timeit(str(selectionsort(lst3)),'from __main__ import selectionsort,lst3')))
print()
print('Time Taken by Bubble Sort to sort 10000 elements: '+str(timeit.timeit(str(bubblesort(lst4)),'from __main__ import bubblesort,lst4')))
print('Time Taken by Insertion Sort to sort 10000 elements: '+str(timeit.timeit(str(insertionSort(lst4)),'from __main__ import insertionSort,lst4')))
print('Time Taken by Selection Sort to sort 10000 elements: '+str(timeit.timeit(str(selectionsort(lst4)),'from __main__ import selectionsort,lst4')))
print()
```
