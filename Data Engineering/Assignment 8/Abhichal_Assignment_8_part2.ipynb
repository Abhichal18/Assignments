i) Fill first column randomly with integers in range [1,100], second column with real numbers in range [-1,1] and third column with alphabets in [a,z].

```python
# creating an empty array of 100*3 shape
data=np.empty((100,3),dtype='O')
# creating an array of alphabets
alphabets=['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q',
          'r','s','t','u','v','w','x','y','z']

# iterating for evry row
for i in range(100):
    # adding first element as a random integer between 1 to 100
    data[i][0]=np.random.randint(1,101)
    # adding second element as a random real values between -1 to 1
    data[i][1]=np.random.uniform(-1,1)
    # adding third element as a random alphabet from a-z
    data[i][2]=np.random.choice(alphabets)
# showing the data
data
```

ii) Convert column three to numeric form by using any convenient mapping to integers.

```python
# iterating for each row
for i in range(100):
    # changing the alphabets to their ascii values in integer form
    data[i][2]=ord(data[i][2])
# showing the data
data
```

iii) Normalize the array obtained in (ii.) using min max normalization. Use normalized array for all later operations.

```python
# normalizing every column separately using min max normalization

# creating a temp variable storing the each row of first column
temp=data[:,0]
# applying the min max normalization on the data
data[:,0]=(temp-temp.min())/(temp.max()-temp.min())

# creating a temp variable storing the each row of second column
temp=data[:,1]
# applying the min max normalization on the data
data[:,1]=(temp-temp.min())/(temp.max()-temp.min())

# creating a temp variable storing the each row of third column
temp=data[:,2]
# applying the min max normalization on the data
data[:,2]=(temp-temp.min())/(temp.max()-temp.min())

#showing the data
data
```

ix) Remove outliers based on the observations in (viii.).

```python
# function for removing the outliers
def remove_outlier(df_in, col_name):
    # finding the 25th quartile
    q1 = df_in[col_name].quantile(0.25)
    # finding the 75th quartile
    q3 = df_in[col_name].quantile(0.75)
    iqr = q3-q1 #Interquartile range
    fence_low  = q1-1.5*iqr
    fence_high = q3+1.5*iqr
    # cleaned data
    df_out = df_in.loc[(df_in[col_name] > fence_low) & (df_in[col_name] < fence_high)]
    return df_out
```

```python
# calling remove outlier for every column except floor and condition
# we didn't call for floor and conditon beause there values get fixed
new_data=remove_outlier(new_data,'price')
new_data=remove_outlier(new_data,'bedrooms')
new_data=remove_outlier(new_data,'bathrooms')
new_data=remove_outlier(new_data,'yr_built')

# showing the cleaned data
new_data
```
