# 2) IMPLEMENT DIJKSTRA'S

Implement Dijkstras algorithm and demonstrate in the above graph to find the diameter of the graph. Diameter of the graph is the longest path among the shortest paths between any two vertices. Draw the graph and highlight the diameter in the graph.

```python
# Importing all Important modules to run the program

# importing networkx to visualize a graph
import networkx as nx
# importing numpy for seeding
import numpy as np
# importing matplotlib for plotting the graph
import matplotlib.pyplot as plt


# this is used to store path for every vertices from source
different_paths=[]
# this is a temporary array which will store tha path for specific index and later apende in the above array
temp=[]


# IMPLEMENTING DIJKSTRA'S ALGORITHMS
def dijkstra(source,graph):
    
    # this have shortest distances from source to every vertex
    distance=[float('inf')]*len(graph)
    
    # distance from source to source is zero
    distance[source]=0
    
    # this contain a tree of shortest path
    short_path=[-1]*len(graph)
    
    # initializing a queue
    queue=[]
    # adding every vertex to queue to iterate on each of them
    for i in range(len(graph)):
        queue.append(i)
    
    # if queue is not empty run the loop
    while queue:
        
        # find the minimum distance neighbor from the vertex which are left in the queue
        u=minimumdistance(distance, queue)
        
        # after getting the indice remove it from the queue
        queue.remove(u)
        
        # this loop will help is updating the distance values and short path subtree
        for v in range(len(graph)):
            
            # checking if there is efge from u to v and v is still in queue or not and distance from source to v is smaller than v or not
            if graph[u][v] and v in queue and distance[v]>distance[u]+graph[u][v]:
                # changing to smaller distance
                distance[v]=distance[u]+graph[u][v]
                short_path[v]=u
    #print(*short_path)
    #print(*distance)'''
    
    return distance, short_path

# this is to find the minimum distance from a vertex to all neighbour nodes, it's used in the main dijkstra function
def minimumdistance(distance,queue):
    
    # intitializing min to maximum value
    minimum=float('inf')
    min_index=-1
    #print(distance,shortest_path)
    
    #iterating over the length of distance
    for i in range(len(distance)):
        # if distance at index i is less than the minimum value we change the value of minimum as well as for min_index
        if distance[i]<minimum and i in queue:
            minimum=distance[i]
            min_index=i
    
    return min_index

# this is a recursive function used to print the paths from the shortest path subtree array
def path(shortest_path,j):
    
    # we are accessing the global arrays for storing paths for each vertex
    global temp, different_paths
    
    # base condition i.e. if source reached stop
    if shortest_path[j]==-1:
        print(j,end=' ')
        temp.append(j)
        different_paths.append(temp)
        temp=[]
        return
    
    # otherwise calling the path function again
    temp.append(j)
    path(shortest_path,shortest_path[j])
    print(j,end=' ')
    
# creating a 2d array to store the graph in the form of adjacency matrix
graph=[[0 for i in range(10)]for j in range(10)]

# using the same graph as in the question 1 but added weights to them for implementation dijkstra
edges=[[1,2,1],[1,3,1],[1,4,1],[1,5,1],[3,4,1],[2,5,1],[4,8,1],[4,6,1],[6,8,1],[5,7,1],[5,9,1],[7,9,1],[8,9,1],[8,10,1],[9,10,1]]

# creating graph with the help of edges and indexing is from 0
for i in range(len(edges)):
    graph[edges[i][0]-1][edges[i][1]-1]=edges[i][2]
    graph[edges[i][1]-1][edges[i][0]-1]=edges[i][2]
#print(len(graph))


print('Adjacency Matrix Representation of graph is as follows: \n')
for i in graph:
    print(*i)

print('\n')

# calling dijkstra function and storing the returned results into two variables
distance,shortest_path=dijkstra(0,graph)

# source is zero
source=0
print('Shortest distance and path from source ',source, ' to other vertices is as follows: \n')
print('\nVertex\t\t\tDistance\t\tPath')
for i in range(1, len(distance)):
    print('\n',source,' --> ',i,'\t\t',distance[i],'\t\t',end='\t')
    path(shortest_path,i)

print('\n')


maximum=float('-inf')

# this will contain the nodes of the longest shortest path 
diameter=[]
#print(different_paths)

#finding the LONGEST SHORTEST PATH from the source with the help of different_paths arrays where we stored the path of every vertex from source
for i in different_paths:
    if len(i)>maximum:
        maximum=len(i)
        diameter=i
#print(diameter)



# NOW PLOTTING THE GRAPH

# initializing all nodes
nodes=[i for i in range(len(graph))]

# adding only edges to implement in networkx
edge=[]
for i in edges:
    edge.append((i[0]-1,i[1]-1))
    
# adding the edges which are on the path of longest shortest path to highlight them
network_edges=[]
for i in range(len(diameter)-1):
    network_edges.append((diameter[i],diameter[i+1]))

    
print('\n---------------------VISUALIZATION----------------------\n')

print('\nThe Highlighted diameter of the graph is picturized below: \n')
# initializing the graph
G=nx.Graph()

# adding nodes to graph
G.add_nodes_from(nodes)
# adding edges to graph
G.add_edges_from(edge)
np.random.seed(123)

# setting spring layout for graph
spring=nx.spring_layout(G)

# definig the size of figure
plt.figure(figsize=(9,7))

# drawing the graph with nodes
nx.draw(G,spring,node_size=700,with_labels=True, font_weight='bold')

# drawing the graph with diameter nodes highlighted
nx.draw_networkx_nodes(G,spring,nodelist=diameter,node_color='r',alpha=0.7)
# drawing the graph with diameter nodes as well as edges highlighted
nx.draw_networkx_edges(G,spring,edgelist=network_edges, width=1,edge_color='r')

# Adding the edge labels
labels={}
for i in edges:
    labels[(i[0]-1),i[1]-1]=i[2]
    
# drawing the graph with labelling the edges with their values
nx.draw_networkx_edge_labels(G, spring, edge_labels=labels,font_size=20)

# showing the graph
plt.show()
```

# 3) IMPLEMENT MERGE SORT AND COMPARE WITH INSERTION SORT

Implement merge sort and demonstrate on a case where merge sort is more efficient than insertion sort. Show the time footprint. (Hint, for clear demonstration you may need a large array).

```python
# Importing important modules which are useful in program

# importing random for generating random integers for the array
from random import randint

# importing time for finding the execution time of merge and insertion sort over the same array
import time

# IMPLEMENTING MERGE SORT
def mergesort(lst):
    
    # if length of array is greater than 1 do the below steps else return element
    if len(lst)>1:
        
        # finding the mid of array
        mid=len(lst)//2
        
        # splitting the array into two parts left and right, i used slicing here 
        left_arr=lst[:mid]
        right_arr=lst[mid:]
        
        # calling mergesort on both the arrays, i.e., left and right array
        mergesort(left_arr)
        mergesort(right_arr)
        
        # intitalizing indexes for left , right and original list to zero for comparison from starting
        i=0
        j=0
        k=0
        
        # run the loop until i and j are less than the length of left and right array respectively
        while i<len(left_arr) and j<len(right_arr):
            
            # if element in left array is less than the right array insert the element of left array at kth position in the list
            if left_arr[i]<right_arr[j]:
                lst[k]=left_arr[i]
                # increment the index of left array
                i+=1
            # if element in right array is less than the left array insert the element of right array at kth position in the list
            else:
                lst[k]=right_arr[j]
                # increment the index of right array
                j+=1
            # incrementing the index of list
            k+=1
        
        # if i is still less than the length of left array copy all elements of it into the list
        while i<len(left_arr):
            lst[k]=left_arr[i]
            i+=1
            k+=1
            
        # if j is still less than the length of right array copy all elements of it into the list
        while j<len(right_arr):
            lst[k]=right_arr[j]
            j+=1
            k+=1
    return lst


# IMPLEMENTING INSERTION SORT
def insertionsort(lst):
    l=len(lst)
    
    # iterating from 2nd index as one element is already sorted
    for i in range(1,l):
        # storing the element at current i into key
        key=lst[i]
        
        # creating the index from which we have to start compairing in backward
        j=i-1
        
        # run the loop until j>0 and key is greater than the value at jth index
        while j>=0 and key<lst[j]:
            
            # change the element position to next
            lst[j+1]=lst[j]
            j-=1
        # it's j+1 because we got out of the loop when element at j is lower, so j+1 will be the correct position
        lst[j+1]=key
    return lst

# intializing two arrays one for sorting with merge sort and other with insertion sort
arr1=[]
arr2=[]

# creating a long array with random integers
for i in range(10000):
    arr1.append(randint(1,10000))
    
# copying elements of arr1 to arr2 with slicing
arr2=arr1[:]
#print(arr1)
print('\n')

# storing the start time before execution
s1=time.time()
# calling mergesort
lst1=mergesort(arr1)
# storing the end time of execution
e1=time.time()
print('Time taken by merge sort to sort the array is: ',e1-s1)
print('\n')
# storing the start time before execution
s2=time.time()
# calling insertion sort
lst2=insertionsort(arr2)
# storing the end time of execution
e2=time.time()
print('Time Taken by insertion sort to sort the array is: ',e2-s2)
```
