# 2) MATRIX MULTIPLICATION

Implement matrix multiplication using list and loop without using numPy or SciPy. 
The code should be general for matrix sizes and user will have a choice to give the matrices as input or choose to go with randomly generated matrices.

```python
# importing random module for generating random integers for matrix reation
from random import randint

print('Press the buttons accordingly:')
print('1: for Manually inputting matrix elements.')
print('2: for random genration of matrix elements.')
pressed_button=int(input())

# checking which button is pressed
if pressed_button==1:
    print('Please Enter the size of matrix1:')
    
    # inputting the row and column size for matrix1 in space separated form
    a,b=map(int, input().split())
    print('Please Enter the size of matrix2:')
    
    # inputting the row and column size for matrix2 in space separated form
    c,d=map(int, input().split())
    
    # checking if no of columns of matrix 1 is equal to no of rows of matrix2, if not multiplication not possible
    if b!=c:
        print('Matrix can not be multiplied as Number of column in matrix1 is not equal to number of rows in matrix2')
    else:
        matrix1=[]
        print('Enter Elements for Matrix1')
        for i in range(a):
            print('Enter Elements of row '+str(i+1)+' of matrix1:')
            
            # inputting the elements of every row in space separated form
            lst=list(map(int, input().split()))
            
            #appending the row in matrix
            matrix1.append(lst)
        matrix2=[]
        print('Enter Elements for Matrix2')
        for i in range(c):
            print('Enter Elements of row '+str(i+1)+' of matrix2:')
            
            # inputting the elements of every row in space separated form
            lst=list(map(int, input().split()))
            
            #appending the row in matrix
            matrix2.append(lst)
        print('\nGiven two matrices are: ',matrix1,matrix2)
        
        # creating an empty matrix which will be used for storing multiplication values of two matrix
        mul_matrix=[[0 for i in range(d)]for j in range(a)]

        # implementing the multiplication procedure of matrix
        for i in range(a):
            for j in range(d):
                for k in range(c):
                    mul_matrix[i][j]+=matrix1[i][k]*matrix2[k][j]
        print('\nMultiplication of given two matrices is: ',mul_matrix)
elif pressed_button==2:
    # generating the size of both the matrices with the help of random function
    row_size_matrix1=randint(1,3)
    col_size_matrix1=randint(1,3)
    # taking same value as previous because if generated random, more probability is there both are different and multiplication is not possible then.
    row_size_matrix2=col_size_matrix1
    col_size_matrix2=randint(1,3)
    
    matrix1=[]
    for i in range(row_size_matrix1):
        lst=[]
        #generating element for each row
        for j in range(col_size_matrix1):
            lst.append(randint(0,1))
        #appending row in the matrix
        matrix1.append(lst)
    matrix2=[]
    for i in range(row_size_matrix2):
        lst=[]
        #generating element for each row
        for j in range(col_size_matrix2):
            lst.append(randint(0,1))
        #appending row in the matrix
        matrix2.append(lst)
    print('\nGenerated Matrices are: ',matrix1,matrix2)
    
    # creating an empty matrix which will be used for storing multiplication values of two matrix
    mul_matrix=[[0 for i in range(col_size_matrix2)]for j in range(row_size_matrix1)]
    
    # implementing the multiplication procedure of matrix
    for i in range(row_size_matrix1):
        for j in range(col_size_matrix2):
            for k in range(row_size_matrix2):
                mul_matrix[i][j]+=matrix1[i][k]*matrix2[k][j]
    
    print('\nMultiplication of randomly generated two matrices is: ',mul_matrix)
```

# 3) PRIORITY QUEUE

Implement a priority queue, where for each element there will be a priority value and higher priority valued will get placed of lower priority values. 
Elements of same priority values should be ordered based on their time of arrival. Use the solutions in Q1 as functions.

```python
# Implementing priority queue with the help of functions made in the question no 1.

# created an empty priority queue with the help of list
priority_queue=list()

total_no_of_operations=int(input('How many operations you want to perform: '))

for _ in range(total_no_of_operations):
    print('Press the buttons accordingly:')
    print('1: for Enqueuing')
    print('2: for Dequeing')
    print('3: for size of priority queue\n')
    pressed_button=int(input())
    
    # checking which button is pressed by the user
    if pressed_button==1:
        element=int(input('Enter element to enque: '))
        priority=int(input('Enter the Priority of element: '))
        
        # calling enqueue function of question 1 to append the element in the priority queue with high priority at higher level
        enqueue(element,priority)
        print('Current priority queue is: ',priority_queue)
    elif pressed_button==2:
        
        # calling dequeue function of question 1 to delete the front element in the priority queue.
        dequeue(priority_queue)
        print('Dequeueing......')
        print('Current priority queue is: ',priority_queue)
    elif pressed_button==3:
        print('The Current size of priority queue is: ',len(priority_queue))
    else:
        print('Invalid Input')
    print('\n')
print('\nFinal priority queue is: ',priority_queue)
        
        
        
```

# 4) UNIQUE WORDS IN A TEXT FILE
Write a program for a general text file and show one output considering this assignment as text data.The program should print a line as follows:

---------

N word1:count1 word2:count2 ... wordN:countN 

---------

where N is the number of unique words in the text file, word:count means word followed 'colon' and then number of times the word has appeared in the input text file. 
Also print a bar graph showing the counts for each word. Use dictionary, list, loop basic data types and basic operations.You can not use text libraries.


```python
# Importing all important modules that will help in writing an executable code

# os is imported to read a file from local computer
import os
# matplotlib is imported to plot a bar graph of the words
import matplotlib.pyplot as plt
import matplotlib
# operator is imported to sort the dictionary based on values
import operator
# OrderedDict is imported to store the sorted dictionary elements orderely
from collections import OrderedDict

# asking the user to provide the address of the input file
file_address=input('Enter the input file address: ')
#assigning the file path to a variable
filename=os.path.abspath(file_address)

# opening a file whose address is given
with open(filename) as f:
    # storing the words of the text file in the list which are splitted at space
    word_list=[word for line in f for word in line.split()]

# creating a dict to store the words and there counts
dict={}
for i in word_list:
    # if word is already in dictionary, increment it's value otherwise append the new word and set it's value to 1
    if i in dict:
        dict[i]+=1
    else:
        dict[i]=1

# creating a new_dict to store the elements of previous dictionary after checking no word is other than the alphabets(i.e. removing the numerals, special characters ,etc)
new_dict={}
for i in dict:
    # checking if word is made from alphabets only
    if i.isalpha():
        # i.lower() is converting the words to lowercase for better efficiency in count
        new_dict[i.lower()]=dict[i]

# printing the unique words followed by the words and there counts separated by colon
print('\n')
print('Output in the required format is as follows: \n')
print(len(new_dict),end=" ")
for i in new_dict:
    print(str(i)+':'+str(new_dict[i]),end=' ')
print('\n')
new_sorted_dict=OrderedDict(sorted(new_dict.items(),key=operator.itemgetter(1),reverse=True))

# creating a dict of 20 most frequent words to display on bar graph as showing all words will gonna create a dense graph which is not readable
new_dict={}
element_count=0
for i in new_sorted_dict:
    # storing elements to show them in bar graph 
    new_dict[i]=new_sorted_dict[i]
    element_count+=1
    # if elemnt count is equal to 20 breaking off
    if element_count==20:
        break
    
#bar_dict={key:ned_dict[key]}

# setting the figure size, so we can easily see the bar graph
plt.figure(figsize=(60,10))
# bar is used to specify elements on x axis
plt.bar(range(len(new_dict)),list(new_dict.values()))
# xticks is used to specify the list xtick location in the graph
plt.xticks(range(len(new_dict)),list(new_dict.keys()))
# changing the font size of x and y axis values
plt.rcParams.update({'font.size':15})

#this is used to show the  bar graph
print('Required Bar Graph is:')
plt.show()
```

# 5) RANDOM WORD GENERATION

Create a list L of characters in English (small and capital both) including punctuation and special characters.

Write a program that will randomly generate a text by taking word from this assignment text randomly and then pick a character in the word randomly and 
substitute that character with a randomly chosen element in L and finally reverse the order of all the characters of that word and print that in a text file.
The program will repeat for N words, where N is the input from the user. The program should also give option of using some other text file as input. 

Example: 'word' is randomly chosen then 'r' is randomly chosen then 'x' randomly chosen from L, then substitute to get 'woxd' and reversing the order we get 'dxow' as output. 
Use dictionary, list, loop basic data types and basic operations. You can not use text libraries.

```python
# Importing all important modules which will be used in this function

# importing random module to generate random integers and to choose an element randomly from a given list
from random import randint
import random
# os is imported to read a file from local computer
import os
# asking the user to provide the address of the input file
file_address=input('Enter the input file address: ')
#assigning the file path to a variable
filename=os.path.abspath(file_address)

# opening a file whose address is given
with open(filename) as f:
    # storing the words of the text file in the list which are splitted at space
    word_list=[word for line in f for word in line.split()]

# creating a dict to store the words and there counts
dict={}
for i in word_list:
    # if word is already in dictionary, increment it's value otherwise append the new word and set it's value to 1
    if i in dict:
        dict[i]+=1
    else:
        dict[i]=1
        
# creating a new_dict to store the elements of previous dictionary after checking no word is other than the alphabets(i.e. removing the numerals, special characters ,etc)
new_dict={}
for i in dict:
    # checking if word is made from alphabets only
    if i.isalpha():
        # i.lower() is converting the words to lowercase for better efficiency in count
        new_dict[i.lower()]=dict[i]

#storing all keys of dictionary(i.e. all unique words of txt file) in alist
txt_file_words=list(new_dict.keys())
# creating a list to store the characters a-z,A-Z and special characters
char_list=list()

print('\nText file words are as follows:\n\n',txt_file_words)
print('\n')
# converting the ascii values to the characters with the help of chr fucntion in python
for i in range(58,127):
    char_list.append(chr(i))
for i in range(32,48):
    char_list.append(chr(i))
print('Character List is as follows: \n\n',char_list)
print('\n')
repeat=int(input('How many times you want to repeat the Program: '))
print('\n')
for _ in range(repeat):
    
    # choosing the random word from the words generated from text file
    random_word=random.choice(txt_file_words)
    #converting the string to list for choosing a character on it as strings are immutable but list are mutable
    random_word=list(random_word)
    #print('Chosen word is: ',random_word)
    
    # choosing the random character from the previously chosen word
    random_word_from_text=random.choice(random_word)
    # choosing the random character from the list of characters and special symbols
    random_word_from_char_list=random.choice(char_list)
    #print(random_word_from_text,random_word_from_char_list)
    
    # changing the ramdomly chosen character from the randomly chosen word with the randomly chosen character from the character list
    for i in range(len(random_word)):
        if random_word[i]==random_word_from_text:
            random_word[i]=random_word_from_char_list
            # once replaced breaking from the loop
            break
            
    # reversing the word to get the final one
    random_word=random_word[::-1]
    
    # converting the list of character into a string to store it in a newly generated text file
    string=''
    for i in random_word:
        #print(i,end='')
        string+=i
        
    # defining the path where new file will be generated
    file=os.path.abspath('new_file.txt')
    
    # appending into the text file the string generated after each iteration
    with open(file,'a+',encoding='utf-8') as fi:
        fi.write(string+' ')

print('\nA text file is generated at defined path. Please check that for generated words.\n')
    
# this is done again here because to show the ouput after leaving two lines blank in text file for clear vision of output
file=os.path.abspath('new_file.txt')
with open(file,'a+',encoding='utf-8') as fi:
    fi.write('\n')
    fi.write('\n')
```
